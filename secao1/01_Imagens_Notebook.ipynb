```python
img_bgr = cv2.imread('animals/00_Connie.jpg')
plt.imshow(img_bgr)
```

A imagem foi carregada corretamente pelo openCV como um array numpy, mas a cor de cada pixel foi classificada como BGR. As funes da Matplotlib espera uma imagem RGB, portanto, para uma exibio correta da imagem,  necessrio trocar esses canais. Essa operao pode ser feita usando as funes de converso openvc cv2.cvtColor () ou trabalhando diretamente com o array numpy.

```python
im_rgb=cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
```

```python
img_gray = cv2.imread('animals/00_Connie.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray,cmap='gray')  #necessrio configurar o parmetro cmap
```

## Redimensionando Imagens

```python
img.shape
```

```python
img_r=cv2.resize(img,(1000,1000))
plt.imshow(img_r)
```

### proporcional

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
img=cv2.resize(im_rgb,(0,0),img,w_ratio,h_ratio)
plt.imshow(img_r)
```

## Espelhando Imagens

```python
plt.imshow(img)
```

```python
# Espelhando em relao ao eixo X
img = cv2.flip(img,0)
plt.imshow(img)
```

```python
# Espelhando em relao ao eixo Y
img = cv2.flip(img,1)
plt.imshow(img)
```

```python
# Espelhando em relao aos dois eixos
img = cv2.flip(img,-1)
plt.imshow(img)
```
